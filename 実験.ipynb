```python
def pick_data(df, pattern_num):
  df = df[['Recording timestamp', 'Presented Stimulus name', 'down', 'up', 'text']]
  data = df.copy()
  data = data.rename(columns={'Recording timestamp':'time_stamp',
                    'Presented Stimulus name':'作業'})
  data['time'] = data['time_stamp'] / 1_000_000
  s = str(pattern_num)
  data = data.query('作業==@s')
  data['time'] = data['time'] - np.min(data['time'])
  data['up'] = data['up'].apply(lambda x: 1 if x==1 else 0)
  data['down'] = data['down'].apply(lambda x: 1 if x==1 else 0)
  data['text'] = data['text'].apply(lambda x: 1 if x==1 else 0)
  data = data.drop(['time_stamp', '作業'], axis=1)
  slide_num = 7
  DF = pd.DataFrame([])
  for i in range(slide_num):
    min_time = 5 + 25*i
    max_time = 25*(i+1)
    tmp = data.query('@min_time <= time <= @max_time')
    tmp['slide_num'] = i + 1
    DF = pd.concat([DF, tmp])

  data = DF
  data = pd.merge(data, pattern_df[[pattern_num, '番号']], left_on='slide_num', right_on=pattern_num)
  data = data.drop(['slide_num'], axis=1)
  group_data = data.groupby('番号').agg({'up': 'sum',
                                      'down':'sum',
                                      'text': 'sum'}).reset_index()
  group_data['sum'] = group_data['up'] + group_data['down'] + group_data['text']
  group_data['ads_rate'] = (group_data['up'] + group_data['down']) / group_data['sum']
  return group_data
```

```python
data1 = pick_data(df1, 1)
data2 = pick_data(df2, 1)
data3 = pick_data(df3, 2)
data4 = pick_data(df4, 2)
data5 = pick_data(df5, 3)
data6 = pick_data(df6, 4)
data7 = pick_data(df7, 4)
data8 = pick_data(df8, 5)
data9 = pick_data(df9, 6)
data10 = pick_data(df10, 7)
data11 = pick_data(df11, 8)
data12 = pick_data(df12, 9)
```
